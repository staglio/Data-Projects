```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# GII 2025 Data
data = {
    'Pillar': [
        'Institutions', 'Human Capital\n& Research', 'Infrastructure', 
        'Market\nSophistication', 'Business\nSophistication'
        ] * 2,
    'Score': [
        # Brazil
        34.2, 37.6, 45.5, 35.4, 36.6,
        # China 
        60.0, 52.0, 64.3, 57.3, 56.0,
    ],
    'Country': ['Brazil'] * 5 + ['China'] * 5
}
```

```python
df = pd.DataFrame(data)
```

```python
plt.style.use("dark_background")
plt.rcParams["font.family"] = "serif"
plt.rcParams["font.serif"] = ["Georgia"]

fig, ax = plt.subplots(figsize=(11, 6))
fig.patch.set_facecolor("#1e1e1e")
ax.set_facecolor("#1e1e1e")

sns.barplot(
    data=df,
    x="Pillar", y="Score", hue="Country",
    palette={"Brazil": "#00c851", "China": "#ff4444"},
    ax=ax
)

# Value labels on top of bars
for bar in ax.patches:
    h = bar.get_height()
    ax.text(
        bar.get_x() + bar.get_width() / 2,
        h + 0.5, f"{h:.1f}",
        ha="center", va="bottom", fontsize=10, fontweight="bold", color="white"
    )

# Titles and axes
ax.set_title(
    "Innovation Indicators — GII 2025 (WIPO)\nBrazil (52nd) vs China (10th)",
    fontsize=14, fontweight="bold", pad=15, color="white"
)
ax.set_xlabel("")
ax.set_ylabel("Score (0–100)", fontsize=11, color="white")
ax.set_ylim(0, 75)
ax.tick_params(colors="white")
ax.yaxis.grid(True, linestyle="--", alpha=0.3)
ax.set_axisbelow(True)
sns.despine()
ax.text(
    0.99, -0.12, "Source: WIPO – Global Innovation Index 2025",
    transform=ax.transAxes, ha="right", fontsize=8, color="gray", style="italic"
)

plt.tight_layout()
plt.savefig("gii_2025_dark.png", dpi=180, bbox_inches="tight", facecolor="#1e1e1e")
plt.show()
```
